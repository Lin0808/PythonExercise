```python
import pandas as pd
import numpy as np
```

```python
data = {"grammer":["Python","C","Java","GO",np.nan,"SQL","PHP","Python"],
       "score":[1,2,np.nan,4,5,6,7,10]}
df = pd.DataFrame(data)
df
```

```python
#select str 'python'
df[df['grammer']=='Python']
```

```python
print(df.columns)
```

```python
#change the name of column 'score' als 'popularity'
df.rename(columns = {'score':'popularity'}, inplace = True)
```

```python
df
```

```python
# count how many times for each grammer
df['grammer'].value_counts()
```

```python
df
```

```python
# fill the NA value with interpolate
df['popularity'] = df['popularity'].fillna(df['popularity'].interpolate())
```

```python
df
```

```python
# select the rows that popularity > 3
df[df['popularity']>3]
```

```python
# remove the dublicate value in grammer
df.drop_duplicates(['grammer'])
```

```python
# caculate the mean value of popularity
df['popularity'].mean()
```

```python
# change the colomun of grammmer into list
df['grammer'].to_list()
```

```python
# save the dataframe als excel
df.to_excel('test.xlsx')
```
